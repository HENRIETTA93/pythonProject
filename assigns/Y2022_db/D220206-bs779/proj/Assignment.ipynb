#### Part 1 – Key maintenance

```python
### Extracting data
import pandas as pd
Ship = pd.read_csv('data/Ships.csv')
Trip = pd.read_csv('data/CLIWOC15.csv')
```

```python
print("Ship: "+str(Ship.shape[0]) + " rows and " + str(Ship.shape[1]) + " columns.")
print("Trip: "+str(Trip.shape[0]) + " rows and " + str(Trip.shape[1]) + " columns.")
```

```python
# drop_duplicates
DimShip = Ship.drop_duplicates()
```

```python
DimShip.info()
```

```python
print("DimShip: "+str(DimShip.shape[0]) + " rows and " + str(DimShip.shape[1]) + " columns.")
```

```python
## The new records that need to be added to the titanicDimension can be found
## by merging the two and picking out those with only a Left Join 
joined = Trip[["ShipName", "ShipType", "Nationality"]].merge(DimShip, on=["ShipName", "ShipType", "Nationality"], how='left', indicator=True)
joined
```

```python
filterd = joined[joined["_merge"] == "left_only"]
filterd = filterd.drop_duplicates()
```

```python
filterd.info()
```

```python
DimShip = DimShip.append(filterd[["ShipName", "ShipType", "Nationality"]],ignore_index=True)
DimShip
```

```python
## Generating a surrogate key
DimShip['Id'] = DimShip.reset_index().index + 1
DimShip[:20]
```

```python
TripExt = Trip.merge(DimShip, on=["ShipName", "ShipType", "Nationality"], how='left')
TripExt
```

#### Part 3 – Dates

```python
### Transformation - Dates
TripMeasures = FactTrip
from datetime import datetime
for i, rec in TripMeasures.iterrows():
    dtstr = str(rec.Year) + '-' + str(rec.Month) + '-' + str(rec.Day)
    TripMeasures.at[i, "DateString"] = dtstr
    try:
        TripMeasures.at[i,"Date"] = datetime.strptime(dtstr, '%Y-%m-%d')
    except:
        pass    
TripMeasures.head(20)
```

#### Extra Credit - Extending the dimension

```python
#  Name1 and Rank1.  Add two columns to the DimShip dataframe and populate them with the corresponding values found in the Trip dataframe
Trip[['Name1','Rank1']]
```

```python
#### Adding columns to a dimension table
## copying column data from Trip 'Name1','Rank1'column to DimShip
#TripExt = Trip.merge(DimShip, on=["ShipName", "ShipType", "Nationality"], how='left')
name1Rank1Info = Trip.loc[:,["ShipName", "ShipType", "Nationality","Name1","Rank1"]]
name1Rank1Info.drop_duplicates(subset=["ShipName", "ShipType", "Nationality"], keep="first", inplace=True)
# This is what a merge would look like ... we will use it to add the Cabin column to the titanicDim dataframe
DimShip.merge(name1Rank1Info, on=["ShipName", "ShipType", "Nationality"], how='left', indicator=True)
```

```python
# Populating the 'Name1','Rank1' field
DimShip["Name1"]=DimShip.merge(name1Rank1Info, on=["ShipName", "ShipType", "Nationality"], how='left', indicator=True)["Name1"]
DimShip["Rank1"]=DimShip.merge(name1Rank1Info, on=["ShipName", "ShipType", "Nationality"], how='left', indicator=True)["Rank1"]
DimShip.head(20)
```

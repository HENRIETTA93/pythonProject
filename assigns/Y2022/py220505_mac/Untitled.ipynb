```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c',s='d' ,data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
iris_df = pd.read_csv("CW_Data.csv")
labels = iris_df['Programme'].unique()
columns = [column
           for column in iris_df.columns
           if column not in ['ID', 'Programme']]  #Get all attributes except id and class labels
 
_, axes = plt.subplots(1, len(columns)) #Prepare plot areas
for c in range(len(columns)):
    sns.histplot(iris_df, x=columns[c], hue="Programme",
                 multiple="stack", ax=axes[c])  #Plot each histogram at the correct area
 

```
